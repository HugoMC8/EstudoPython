```python
email = "usuario.88@gmail.com"
nome = "usuario"
print (len(email))
print (len(nome))
# pro python o primeiro caracter e contado como indice 0
print (email[0])
print (email[1])
print (email[2])
print (email[3])
print (email[4])
print (email[5])
print (email[6])

#tambem podemos pegar um indice negativo

print (email[-3])
print (email[-2])
print (email[-1])
print (email[-20])

#tambem podemos pegar um pedaço do texto atraves do indice

print (email[:1])
print (email[:4])
print (email[:-1])
print (email[:-4])
#se colocar o indice depois dos ":" voce vai até aquele inidice incluindo ele
#se colocar o indice antes dos ":" voce vai apartir daquele inidice não incluindo ele

print (email[1:])
print (email[4:])
print (email[-1:])
print (email[-4:])
```

 Formulas e métodos de String
 
 texto . metodo() (sera sempre assim, texto ponto final, seguido do método e abre e fecha parenteses
 


```python
#captalize() coloca a 1 letra maiuscula
texto = "hugo cavalcante"
print (texto.capitalize())
texto = "HuGo CaVaLcAnTe"
print (texto.capitalize())
texto = "hUGO CAVALCANTE"
print (texto.capitalize())

#title() coloca a 1 letra de cada palavra em maiuscula

texto = "hugo cavalcante"
print (texto.title())
texto = "HuGo CaVaLcAnTe"
print (texto.title())
texto = "hUGO CAVALCANTE"
print (texto.title())


#casefold() transforma todas letras em minusculas
texto = "HUGO CAVALcante"
print (texto.casefold())
texto = "HuGo CaVaLcAnTe"
print (texto.casefold())

#upper() transforma todas letras em maiusculas
texto = "HUGO CAVALcante"
print (texto.upper())
texto = "HuGo CaVaLcAnTe"
print (texto.upper())


#count() retorna a quantidade de vezes que o valor aparece na string

texto = "HuGo CaVaLcAnTe"
print (texto.count("a"))#case sensitive; pu seja conferir se precisa fazer uma transformação antes.
print (texto.count("A"))
texto = texto.casefold()
print (texto.count("a"))
print (texto.count("A"))

#endswith() verifica se o texto termina com o valor específico e da como resposta true ou false

texto = "usuario.88@gmail.com"
print(texto.endswith(".com"))

#starswith( verifica se o texto começa com o valor expecífico e da como resposta true ou false)
texto = "usuario.88@gmail.com"
print(texto.endswith("usuario"))


#find() procura um texto dentro de outro texto e da como resposta a posicao (indice)
texto = "usuario.88@gmail.com"
print(texto.find("@"))

#isalnum() verifica se o texto e alfa numerico
texto = "123ABC"
print(texto.isalnum())
texto = "123ABC*" #o caracter especial nao e aceito
print(texto.isalnum())

#isalpha() verifica se o texto e apenas letras
texto = "ABC"
print(texto.isalpha())
texto = "ABC123"
print(texto.isalpha())


#isnumeric() verifica se o texto e apenas numeros
texto = "123"
print(texto.isnumeric())
texto = "ABC123"
print(texto.isnumeric())


#replace(A,B) substirui um texto por outro texto em uma string

texto = "1000.00"
print(texto.replace(".",","))


#split() separa uma string em mais de 1 de acordo com o delimitador.

texto = "usuario.88@gmail.com"
print(texto.split("@"))
texto = "usuario.88@gmail.com.br"
print(texto.split("."))

#splitlines() separa um texto em varios textos de acordo com os "enters" do texto

texto = """ola, bom dia
venho por meio desse informar que seu boleto encontra-se em atraso
o valor total e de 1000,00"""
print (texto)
print(texto.splitlines())

#strip() retira caracteres de espaço no inicio e no final do texto
texto = " usuario   "
print(texto.strip())
```
